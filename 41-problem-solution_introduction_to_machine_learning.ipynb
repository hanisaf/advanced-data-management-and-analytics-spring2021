```python
import pandas as pd
from sklearn import tree
from sklearn.tree import export_text
import altair as alt
from vega_datasets import data
```

```python
iris = data.iris()
iris
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalWidth',
    y='petalLength',
    color='species'
) | alt.Chart(iris).mark_point().encode(
    x='sepalWidth',
    y='sepalLength',
    color='species'
)
```

Using `DecisionTreeClassifier` fit a decision tree to the data

- What are the attributes/features?
- What is the target feature?
- Interpret the tree

```python
dtree = tree.DecisionTreeClassifier().fit(
    iris[['petalWidth','petalLength','sepalWidth','sepalLength']], iris['species'])
```

```python
print(export_text(dtree, feature_names=['petalWidth','petalLength','sepalWidth','sepalLength']))
```
