## Data Description

### Surveys Data
- `record_id`: Unique id for the observation
- `month`: month of observation
- `day`: day of observation
- `year`: year of observation
- `plot_id`: ID of a particular site
- `species_id`: 2-letter code
- `sex`: sex of animal (“M”, “F”)
- `hindfoot_length`: length of the hindfoot in mm
- `weight`: weight of the animal in grams

### Cars Data
- `mpg`: Miles/(US) gallon
- `cyl`: Number of cylinders
- `disp`: Displacement (cu.in.)
- `hp`: Gross horsepower
- `drat`: Rear axle ratio
- `wt`: Weight (1000 lbs)
- `qsec`: 1/4 mile time
- `vs`: V/S
- `am`: Transmission (0 = automatic, 1 = manual)
- `gear`: Number of forward gears
- `carb`: Number of carburetors

````python
# import packages such as as plotnine
import plotnine as p9
import pandas as pd
````

````python
# upload sample data
surveys_complete = pd.read_csv('surveys.csv').dropna().reset_index(drop=True)
surveys_complete.
````

````python
surveys_sub = surveys_complete.loc[surveys_complete['species_id'] == 'RM']
````

## Load Complete, Graphing Begins
We have loaded in our plotnine package and our surveys data on animals of multiple species recording their hindgoot length and weight. The surveys sub file hones in on a specific species of animal labelled `RM`. We will use these sets to plot visualizations based on the commented example below. 
- Select the dataset
- Map which elements will make the x and y fields
- choose what type of graph (point, bar, line, ...) 
- Add any extra features you may desire.

````python
# (p9.ggplot(data = ...,
#            mapping = p9.aes(x = ..., y = ...))
#     + p9.geom_point()
# )
````

````python
# Following the model above, we make the two elements necessary for our visual: the mapping and the plot on top
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point()
)
````

````python
# Create
surveys_plot = p9.ggplot(data=surveys_complete,
                         mapping=p9.aes(x='weight', y='hindfoot_length'))

# Draw the plot
surveys_plot + p9.geom_point()
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='plot_id'))
    + p9.geom_bar()
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='species_id', y = ''))
    + p9.geom_bar()
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point(alpha=0.1)
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
    + p9.geom_point(alpha=0.1, color='blue')
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha = 0.5)
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight',
                          y='hindfoot_length',
                          color='species_id'))
    + p9.geom_point(alpha=0.1)
)
````

````python
(p9.ggplot(data = surveys_sub,
        mapping = p9.aes('weight', 'hindfoot_length', color='sex')) +
        p9.geom_point() + 
        p9.stat_smooth(method='lm')
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
    + p9.geom_point(alpha=0.1)
    + p9.xlab("Weight (g)")
    + p9.scale_x_log10()
)
````

````python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='plot_id',
                          fill='sex'))
    + p9.geom_bar()
    + p9.scale_fill_manual(["blue", "orange"])
)
````

````python
yearly_counts = surveys_complete.groupby(['year', 'species_id'])['species_id'].count()
yearly_counts = yearly_counts.reset_index(name='counts')
yearly_counts
````

````python
(p9.ggplot(data=yearly_counts,
           mapping=p9.aes(x='year',
                          y='counts',
                          color='species_id'))
    + p9.geom_line()
)
````
