```python
#pip install graph_tools
import graph_tools as gt
from graph_tools import Graph
```

```python
#create empty graph by instantiating Graph class
graph = Graph()

#create empty undirected graph with value parameter
ugraph = Graph(directed = False)
```

```python
#populate graph with edges
ugraph.add_edge(1,2)
ugraph.add_edge(1,3)
ugraph.add_edge(1,6)
ugraph.add_edge(2,3)
ugraph.add_edge(3,4)
ugraph.add_edge(4,5)
ugraph.add_edge(4,7)
```

```python
#populate graph with nodes
ugraph.add_vertex(8)
ugraph.add_vertex(9)
```

```python
#delete specified edge
ugraph.delete_edge(1,6)
```

```python
#delete specified node
ugraph.delete_vertex(9)
```

```python
#set weights to nodes
ugraph.set_vertex_weight(2,4)
ugraph.set_vertex_weight(3,6)
ugraph.set_vertex_weight(4,2)
ugraph.set_vertex_weight(7,1)

print(ugraph)
```

```python
#return the average degree of all nodes
ugraph.average_degree()
```

```python
#check betweenness of a node
ugraph.betweenness(3)
```

```python
#create adjacency matrix
ugraph.adjacency_matrix()
```

```python
#create diagonal matrix
ugraph.diagonal_matrix()
```

```python
#create barabasi graph with 15 nodes
bb = ugraph.copy_graph()
bbgraph = bb.create_barabasi_graph(N = 15)
print(bb)
```
