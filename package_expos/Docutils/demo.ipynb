```python
import docutils
from docutils import core, io
```

<h2> rst file to XML </h2>

Code below from this website: [https://www.kite.com/python/docs/docutils] 

```python
#This code converts the rst file to a XML file and prints it out
docutils.core.publish_file(
    source_path="trial.rst",
    destination_path="output2.xml",
    writer_name="xml")
    
result = open("output2.xml").read()
print(result)
```

<h2> HTML file to XML </h2>


HTML code created with HTML tags

```python
#This code converts the html file to a XML file and reads it out to view
docutils.core.publish_file(
    source_path="outputHTML.html",
    destination_path="output.xml",
    writer_name="xml")

xml = open("output.xml").read()
print(xml[7316:])
```

<h2> rst file to HTML </h2>

Code below from this website: [https://github.com/marianoguerra/rst2html5/blob/master/README.rst] 

```python
#This code converts the rst file to a HTML file and reads it out to see below
docutils.core.publish_file(
    source_path="trial.rst",
    destination_path="outputHTML.html",
    writer_name="html")

html = open("outputHTML.html").read()
print(html[7316:])
```

<h2> rst file to LaTex file </h2>

Code below from this website: [https://www.kite.com/python/examples/4102/docutils-convert-a-restructuredtext-file-to-latex] 

```python
#This code converts the rst file to a LaTex file and reads it out to view
import docutils.core
docutils.core.publish_file(
    source_path="trial.rst",
    destination_path="output.tex",
    writer_name="latex")
    
result = open("output.tex").read()
print(result[749:])
```
