## Basic Overview
In this expo we will showcase how Seaborn is able to produce visualizations for the following:
* Data Distribution
* Categorical Data
* Heatmaps
* Regression Plots

## distplot

The first data distribution related visualization we will look at is distplot which shows the distribution of a univariate set of observations.

```python
sns.distplot(car['total'])
```

## jointplot

Our next visualization jointplot() allows you to visualize a plot of two variables with bivariate and univariate graphs. The frequency distributions of each variable are shown on the far side of each axis.

```python
sns.jointplot(x='total',y='not_distracted',data=car,kind='scatter')
```

```python
sns.jointplot(x='total',y='not_distracted',data=car,kind='hex')
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns). This is exceptionally useful in helping discover relationships between variables. The plots running along the diagonal are histograms of the corresponding variable.

```python
sns.pairplot(car)
```

## barplot

Now let's introduce some methods of visualizing categorical data. barplot allows you to aggregate categorical data based off some function. If a function is not given, barplot will default the mean:

```python
sns.barplot(x='origin',y='mpg',data=mpg)
```

### countplot

countplot is a simplified version of barplot where we aggregate on the categorical data, but the graph only counts the number of observations within that category.

```python
sns.countplot(x='origin',data=mpg)
```

## boxplot

boxplot is used to show a detailed look at the distribution of categorical data. It is more commonly refered to as a box-and-whisker plot.

```python
sns.boxplot(x="origin", y="mpg", data=mpg)
```

## Heatmap

Data used for a heatmap needs to be in matrix form. Here we will create a heatmap using correlation data from our dataset.

```python
# Matrix form for correlation data
car.corr()
```

```python
sns.heatmap(car.corr())
```

# lmplot

The last functionality of Seaborn we will showcase is the function **lmplot**, which allows us to display linear models. There are added functionalities to lmplot that allows you split up the plots based off of features, as well as coloring the hue based off of features.

```python
sns.lmplot(x='not_distracted',y='total',data=car)
```
