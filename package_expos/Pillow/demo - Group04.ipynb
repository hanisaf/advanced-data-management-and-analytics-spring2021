### Cutting, Pasting, and Merging

```python
box = (150, 5, 300, 110)
```

```python
region = im.crop(box)
```

```python
region
```

```python
region = region.transpose(Image.ROTATE_180)
```

```python
region
```

```python
im1 = Image.open("uga.jpg")
```

```python
im1.paste(region, box) 
```

```python
im1
```

```python
im.getbands()
```

```python
r, g, b = im.split()
```

```python
Image.merge("RGB", (b, g, r))
```

### Color transformation

```python
with Image.open("uga.jpg") as im:
    im = im.convert("L")
```

```python
im
```

```python
with Image.open("uga.jpg") as im:
    im = im.convert("RGB")
```

```python
im
```

```python
from PIL import ImageFilter
im.filter(ImageFilter.DETAIL) 
```

```python
from PIL import ImageEnhance
```

```python
enh = ImageEnhance.Contrast(im)
enh.enhance(1.3).show("30% more contrast")
```
