## Explicitly create figures and axes, and call methods ("object-oriented (OO) style")

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  #Create a figure and an axes
ax.plot(x, x, label='linear')  #Plot some data
ax.plot(x, x**2, label='quadratic')  
ax.plot(x, x**3, label='cubic') 
ax.set_xlabel('x label')  #Add an x-label 
ax.set_ylabel('y label')  #Add a y-label 
ax.set_title("Simple Plot")  #Add a title 
ax.legend()  #Add a legend
```

## Pyplot automatically creates and manages the figures and axes, and uses pyplot functions

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

## Import the data from vega_datasets 'cars'

```python
data = data.cars()
data.head()
```

```python
#These are the X, Y values we will be using from the entire unedited dataset
MPG = data['Miles_per_Gallon']
Horsepower_data = data['Horsepower']
```

### Create a subset dataframe for different visualization capabilities

```python
df = pd.DataFrame(data.groupby('Year')['Horsepower'].mean())
df
```

```python
#These are the X, Y values we will be using from df
Year = df.index
Horsepower = df['Horsepower']
```

## Create another subset dataframe for different visualization capabilities

```python
df2 = pd.DataFrame(data.groupby('Cylinders')['Acceleration'].mean())
df2
```

```python
#These are the X, Y values we will be using from df2

Cylinders = df2.index
Acceleration = df2['Acceleration']
```

### Creating a simple line graph

```python
plt.plot(Year, Horsepower)
plt.xlabel("Year")
plt.ylabel("Horsepower")
```

### Adjusting line settings

```python
plt.plot(Year, Horsepower, color = 'blue', linewidth =  5)
```

### Creating a bar chart

```python
plt.bar(MPG, Horsepower_data)
```

```python
#bar chart using cylinders and acceleration
plt.bar(Cylinders, Acceleration)
```

### Creating a histogram

```python
plt.hist(MPG, bins = 10)
```

### Creating a scatterplot

```python
plt.scatter(MPG, Horsepower_data)
```

```python
#adding color
plt.scatter(MPG, Horsepower_data, color = 'red')
```

```python
#adding axis labels & a title
plt.scatter(MPG, Horsepower_data, color = 'red')
plt.title('How MPG Affects Horsepower')
plt.ylabel('Horesepower')
plt.xlabel('Miles per Gallon')
```

### Now we can play with a few of the line settings from the table above!

```python
plt.plot(Year, Horsepower, color = 'red', linestyle = '--')
```

```python
plt.plot(Year, Horsepower, color = 'red', linestyle = ':') 
#other linestyles: '-', '--',':','-.'
```

```python
plt.plot(Year, Horsepower, color = 'red', linestyle = '--', linewidth = 0.5)
```

```python
plt.plot(Year, Horsepower, color = 'red', linestyle = '--', linewidth = 0.5, marker = 'o')
```

## Creating a final subset dataframe to see how Matplotlib handles categorical data

```python
df3 = pd.DataFrame(data.groupby('Origin')['Miles_per_Gallon'].mean())
df3
```

```python
#These are the X, Y values we will be using from df3

Orgin = df3.index
MPG_Origin = df3['Miles_per_Gallon']
```

```python
plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(Orgin, MPG_Origin)

plt.subplot(132)
plt.scatter(Orgin, MPG_Origin)

plt.subplot(133)
plt.plot(Orgin, MPG_Origin)

plt.suptitle('Categorical Plotting')
plt.show()
```
